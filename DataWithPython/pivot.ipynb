```python
# For this we will use a dataset from seabornÂ 
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
flights = sns.load_dataset('flights')
flights.head(14)
```

```python
flights.groupby('year').sum('passengers')
```

```python
flights_pivoted=pd.pivot_table(flights,
                            values='passengers',
                            index=['year'],
                            columns='month',
                            aggfunc=np.sum
)
flights_pivoted
```

```python
flights_pivoted =pd.pivot_table(flights,values="passengers",index=["year"],columns="month",aggfunc=np.sum)
flights_pivoted
```

```python
flights_pivoted = pd.pivot_table(flights,
                                values='passengers', # Values in the table
                                index=['year'],  # the x axis
                                columns='month', # The y axis
                                aggfunc=np.sum).plot(kind='bar',width = 0.8, edgecolor='black')
```

```python
table= pd.pivot_table(flights_pivoted, values='D', index=['A', 'B'],
                        columns=['C'], aggfunc=np.sum)
```
