```python
tips = sns.load_dataset('tips')
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=5, bins=30, kde=True)
```

```python
sns.histplot(data=tips, x="size", stat="percent", color="navy", discrete=True)
```

```python
sns.histplot(data=tips, x="total_bill", binwidth=5, bins=30, kde=True, hue='smoker',palette='magma', multiple="stack")
```
